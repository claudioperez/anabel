```python
import ema as em
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format = 'svg' # used to make plots look nicerbb
```

# Problem 2
(See also HW-07 Problem 1)

```python
E = 1000
At = 30
A = 30000000
I = 100
```

```python
#Remove
mdl = em.rModel(2,3)
n = mdl.dnodes
e = mdl.delems

mdl.material('default', E)
mdl.xsection('default', A, I)
truss_section = mdl.xsection('truss', At, I)

mdl.node('1',  0.0, 0.0)
mdl.node('2',  8.0, 0.0)
mdl.node('3',  8.0, 6.0)
mdl.node('4', 16.0, 6.0)

mdl.beam('a', n['1'], n['2'])
mdl.beam('b', n['2'], n['3'])
mdl.beam('c', n['3'], n['4'])
mdl.truss('d', n['2'], n['4'], xsec=truss_section)

mdl.hinge(e['c'], n['4'])

mdl.fix(n['1'], ['x', 'y', 'rz'])
mdl.fix(n['4'], ['y', 'rz'])

mdl.DOF = [[5, 6, 7], [5, 1, 2], [3, 1, 4], [3, 8, 9]]
```

```python
fig, ax = plt.subplots(1,1)
em.plot_structure(mdl, ax)
```

### Find global displacements

```python
K = em.K_matrix(mdl)
K.f
```

```python
B = em.B_matrix(mdl)
Q0 = em.Q_vector(B).o
Q0
```

```python
P0 = B.f@Q0
P0
```

```python
P = np.array([-25, 0., 0., -40])
Uf = K.f.inv@ P
```

```python
Uf
```

```python
Uf = em.analysis.SolveDispl(mdl)
Uf
```
