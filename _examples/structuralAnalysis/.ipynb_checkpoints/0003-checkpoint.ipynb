```python
import ema as em
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

```python
mdl = em.Model(2,3)
n = mdl.dnodes
e = mdl.delems

mdl.material('default', E=1000)
mdl.xsection('default', 1e6, 100)
xt = mdl.xsection('default', 10, 1)

mdl.node('1', 0.0, 0.0)
mdl.node('2', 0.0, 6.0)
mdl.node('3', 4.0, 6.0)
mdl.node('4', 8.0, 6.0)
# mdl.node('5',  0.0, 6.0)

mdl.beam('a', n['1'], n['2'])
mdl.beam('b', n['2'], n['3'])
mdl.beam('c', n['3'], n['4'])
mdl.truss('d', n['1'], n['4'], xsec=xt)

mdl.fix(n['1'], ['x','y'])
mdl.fix(n['4'], ['y'])

mdl.numDOF()
```

```python
fig, ax = plt.subplots()
em.plot_structure(mdl, ax, labeled=True)
```

```python
e['a'].e0['2'] = 2e-3
e['b'].e0['2'] = 2e-3
e['c'].e0['2'] =-4e-3
```

```python
Uf = em.analysis.SolveDispl(mdl)
Uf
```

```python
fig, ax = plt.subplots()
em.plot_U(mdl, Uf, ax, scale=10)
```
