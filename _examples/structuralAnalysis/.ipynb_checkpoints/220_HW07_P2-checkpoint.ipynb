```python
import ema as em
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

# Problem 2

```python
#Remove
P2 = em.Domain(2,3)
n = P2.dnodes
e = P2.delems

P2.node('1', 0.0, 0.0)
P2.node('2', 0.0, 6.0)
P2.node('3', 8.0, 6.0)
P2.node('4', 8.0, 2.0)
P2.node('5', 14., 2.0)

P2.beam('a', n['1'], n['2'])
P2.beam('b', n['2'], n['3'])
P2.beam('c', n['3'], n['4'])
P2.beam('d', n['4'], n['5'])
P2.truss('e', n['1'], n['3'])

P2.pin(n['1'])
P2.fix(n['5'], ['x', 'y', 'rz'])

P2.DOF = [[11,12,8],[1,9,2],[6,3,4],[10,7,5],[13,14,15]]
```

```python
#Remove
fig, ax = plt.subplots(1,1)
em.plot_structure(P2, ax)
```

```python
A = em.A_matrix(P2).remove("frame-axial").add_cols(['1', '6']).add_cols(['3', '7']).add_cols(['12', '9']).add_cols(['13', '10'])
U = em.U_vector(A)
V = em.V_vector(A)
```

```python
B = em.B_matrix(P2).add_rows(['1', '6']).add_rows(['3', '7'])
B = B.add_rows(['13', '10']).add_rows(['12', '9'])
Bf = B.f.del_zeros()
Qf = em.Q_vector(Bf)
Bf.df
```

```python
V.set_item('a_2', -0.6937e-3)
V.set_item('a_3',  1.3874e-3)
V.set_item('b_2',  0.0765e-3)
V.set_item('b_3', -2.9278e-3)
V.set_item('c_2', -0.1310e-3)
V.set_item('c_3',  3.1515e-3)
V.set_item('d_2', -3.5215e-3)
V.set_item('d_3', -2.2149e-3)
V.set_item('e_1',  6.2922e-3)

Qf.set_item('a_3', 27.748)
Qf.set_item('b_2',-27.748)
Qf.set_item('b_3',-57.791)
Qf.set_item('c_2', 57.791)
Qf.set_item('c_3', 123.440)
Qf.set_item('d_2',-123.440)
Qf.set_item('d_3',-106.019)
Qf.set_item('e_1', 12.584)
```
