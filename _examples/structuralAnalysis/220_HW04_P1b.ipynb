```python
import numpy as np
import sympy as sp
import ema as em
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

# Problem 1

```python
domain = em.Domain(2,3)
m1, s1 = em.UnitProperties()

n1 = domain.node('1',  0.0, 8.0)
n2 = domain.node('2', 10.0, 8.0)
n3 = domain.node('3', 20.0, 8.0)
n4 = domain.node('4',  0.0, 0.0)
n5 = domain.node('5', 10.0, 0.0)

a = domain.beam('a', n1, n2, m1, s1)
b = domain.beam('b', n2, n3, m1, s1)
c = domain.beam('c', n2, n5, m1, s1)
d = domain.beam('d', n4, n5, m1, s1)

domain.hinge(b, n2)
domain.hinge(d, n4)
domain.hinge(d, n5)

domain.pin(n1)
domain.fix(n4, ['x', 'y', 'rz'])
domain.fix(n3, 'y')

domain.numDOF()
```

```python
fig, ax = plt.subplots( 1, 1)
em.plot_structure(domain, ax)
```
