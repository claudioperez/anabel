```python
mdl = em.rModel(2,3)
n = mdl.dnodes
e = mdl.delems
mdl.material('default', 1000)
mdl.xsection('default', 1000000000, 50)
xt = mdl.xsection('truss', 20, 1)

mdl.node('1',  0.0, 0.0)
mdl.node('2',  8.0, 0.0)
mdl.node('3',  8.0, 6.0)
mdl.node('4', 16.0, 6.0)

mdl.beam('a', n['1'], n['2'])
mdl.beam('b', n['2'], n['3'])
mdl.beam('c', n['3'], n['4'])
mdl.truss('d', n['2'], n['4'], xsec=xt)

mdl.hinge(e['a'], n['1'])

mdl.fix(n['1'], ['y', 'rz'])
mdl.fix(n['4'], ['x', 'y', 'rz'])

mdl.DOF = [[1, 5, 6],[1, 2, 3],[7, 2, 4],[7, 8, 9]]
```

```python
#Remove
fig, ax = plt.subplots()
em.plot_structure(mdl, ax)
```
