```python
import ema as em
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format = 'svg'
```

```python
dm = em.Domain(2,3)
e = dm.delems
n = dm.dnodes

dm.node('1',  0.0, 0.0)
dm.node('2',  4.0, 0.0)
dm.node('3',  7.0, 0.0)
dm.node('4',  4.0, -4.0)
dm.node('5',  4.0, 4.0)


dm.beam('a', n['1'], n['2'])
dm.beam('b', n['2'], n['3'])
dm.beam('c', n['2'], n['4'])
dm.beam('d', n['2'], n['5'])
dm.truss('e', n['3'], n['5'])

dm.hinge(e['c'], n['4'])

dm.fix(n['1'], ['y'])
dm.fix(n['4'], ['x', 'y', 'rz'])
dm.numDOF()
```

```python
fig, ax = plt.subplots()
em.plot_structure(dm, ax)
```
