```python
import ema as em
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

# Problem 4

## Geometry

```python
domain = em.Domain(2,3)
elems = domain.elems

n1 = domain.node('1',  0.0,  0.0)
n2 = domain.node('2',  8.0,  0.0)
n3 = domain.node('3',  8.0,  6.0)
n4 = domain.node('4', 16.0,  6.0)
n5 = domain.node('5', 16.0, -4.0)

a = domain.beam('A', n1,  n2)
b = domain.beam('B', n2,  n3)
c = domain.beam('C', n3,  n4)
d = domain.beam('D', n4,  n5)
e = domain.truss('E', n2,  n4)

domain.hinge(a, n1)
domain.hinge(d, n5)

domain.fix(n1, ['x', 'y', 'rz'])
domain.fix(n5, ['x', 'y', 'rz'])


fig, ax = plt.subplots(1,1)
em.plot_structure(domain, ax)
```

```python
domain.numDOF()
```

```python
domain.redundant(a, '3')
domain.redundant(c, '2')
```

```python
B = em.B_matrix(domain)
# combine dofs colinear with frame axial forces
Br = B.add_rows(['2', '5']).add_rows(['4','7'])

# combine dofs colinear with rxns
Br = Br.add_rows(['10', '1']).add_rows(['14','8'])
# Br = Br.f.del_zeros()
Br = Br.f
Br
```
