```python
import ema as em
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

# Problem 3

```python
P3 = em.Domain(2,3)
P3.n = P3.dnodes
P3.e = P3.delems

P3.node('1', 0.0, 0.0)
P3.node('2', 0.0, 6.0)
P3.node('3', 4.0, 6.0)
P3.node('4', 8.0, 6.0)

P3.beam('a',  P3.n['1'], P3.n['2'])
P3.beam('b',  P3.n['2'], P3.n['3'])
P3.beam('c',  P3.n['3'], P3.n['4'])
P3.truss('d', P3.n['1'], P3.n['4'])

P3.hinge(P3.e['b'], P3.n['3'])
P3.hinge(P3.e['a'], P3.n['1'])

P3.fix(P3.n['1'], ['x', 'y', 'rz'])
P3.fix(P3.n['4'], 'y')

P3.numDOF();
```

```python
fig3, ax3 = plt.subplots(1,1)
em.plot_structure(P3, ax3)
```

## Problem 4

```python
p4 = em.Domain(2,3)
e = p4.delems # reference to dictionary of elements
n = p4.dnodes # reference to dictionary of nodes

p4.node('1',  0.0, 0.0 )
p4.node('2', -6.0, 6.0)
p4.node('3',  0.0, 6.0)
p4.node('4',  6.0, 6.0)
p4.node('5',  0.0, 14.0)

p4.beam('a', n['1'], n['3'])
p4.beam('b', n['2'], n['3'])
p4.beam('c', n['3'], n['4'])
p4.beam('d', n['3'], n['5'])

p4.truss('e', n['2'], n['5'])
p4.truss('f', n['4'], n['5'])

p4.hinge(e['b'], n['3'])
p4.hinge(e['d'], n['3'])

p4.fix(n['1'], ['x', 'y', 'rz'])

p4.numDOF();
```

```python
fig, ax = plt.subplots(1,1)
em.plot_structure(p4, ax)
```

### $B$ Matrix

```python
B = em.B_matrix(p4)
B.df
```

### $A$ Matrix

```python
A = em.A_matrix(p4)
A.df
```
