```python
import jax
import anon
import anon.atom as anp
from anon import quad
import matplotlib.pyplot as plt
from elle_0001 import elem_0001
```

```python
f = lambda x: anp.pi**2/100*sum(anp.sin((2*k+1)*anp.pi*x) for k in range(5))
```

```python
elem = elem_0001(f,1,1,_expose_closure=True)
transf = elem.closure["transf"]
grad_transf = elem.closure["grad_transf"]
```

```python
fix,ax = plt.subplots(1,2)
for el,rng in enumerate([anp.linspace(0,1,5), anp.linspace(2,10,5)]):
    for xi in anp.linspace(-1,1,10):
        ax[el].plot([xi,transf(xi,rng)], [0,0.5])
        ax[el].plot(*zip(*[[xi,grad_transf(xi,rng)] for xi in anp.linspace(-1,1,10)]))
```
