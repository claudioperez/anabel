```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("typewriter")
SAVE_FIGS=True
```

```python
RHO = 0
M = 1
def U(**kwds):
    u = [None,None]
    u[RHO] = kwds['rho']
    u[M] = kwds['m']
    return u

def u1(u, xi,a=1):
    rho = u[RHO]*(xi+1)
    m = u[M] + xi*(u[M]-a*u[RHO]*np.sqrt(1+xi))
    return rho, m

def u2(u, xi,a=1):
    rho = u[0]*(xi+1)
    m = u[1] + xi*(u[1]+a*u[0]*np.sqrt(1+xi))
    return rho, m
u = u1,u2
```

```python
fig,ax = plt.subplots()
rho = (1,)
m = (0,3)
for i in range(2):
    XI = np.linspace(-1,3.5,100)
    u0 = U(rho=rho[0],m=m[i])
    ax.scatter([u0[0]],[u0[1]],marker="x",color="rk"[i])
    for p in (0,1):
        ax.plot(*zip(*[u[p](u0,xi) for xi in XI ]),":-"[p],color="rk"[i],label=f"$\\mathbf{{u}}_{'02'[i]}$ {p+1}-shock")
ax.legend()
ax.set_title("Hugoniot Locus")
ax.set_xlabel("$\\rho(\\xi)$")
ax.set_ylabel("$m(\\xi)$")
if SAVE_FIGS: fig.savefig("img/p1-locus.png")
```
